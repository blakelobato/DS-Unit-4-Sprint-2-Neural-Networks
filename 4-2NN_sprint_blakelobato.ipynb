## 1. Defining Neural Networks 

Write *your own* definitions for the following terms:

- **Neuron:**
    - A neuron is the equivalent to a neuron within the brain that sends signals throughout the brain. It is the basic unit for a NN. These neurons are basically nodes that receive inputs applies a weight and bias value, then utilizes an activation function in order to pass on information to the next layer of neurons in the network. 
    
    
- **Input Layer:**
    - An input layer is the first layer of a Neural Network. It recieves inputs or values from the data in question and begins working through the dataset once receiving inputs.
    
    
- **Hidden Layer:**
    - The hidden layer is the layer or layers behind the input layer. This layer receives the input values and calculations that were passed on from the input layer. Here the sum of inputs are multiplied by weights and a bias is added to them. There can be multiple hidden layers if need be to find a better solution. 
    
    
- **Output Layer:**
    - The output layer is the result of the model or output after all the data has been passed through the input layer and hidden layer(s). These represent the final result that will be used. 
    
    
- **Activation Function:**
    - The activation function is used to reduce some high variability between data values. Sigmoid functions, relu, prelu, tanh, etc. are all examples of activation functions that take in the values and compress them into a range (e.g. sigmoid functions squish the values between 0 and 1 in order to eliminate values that could be 5 or 1000 thus introducing non-linearity). These functions help to determine the amount of 'signal' is transmitted through each layer.

```python
def create_model(init_mode='uniform'):
    # create model
    model = Sequential()
    model.add(Dense(32, kernel_initializer=init_mode, input_shape=(inputs,), activation='relu')) #input layer 
    model.add(Dense(100, kernel_initializer=init_mode, activation='relu')) #hidden layers
    model.add(Dense(8, kernel_initializer=init_mode))
    model.add(tensorflow.keras.layers.PReLU()) #add a prelu activation to the model
    model.add(Dense(1, kernel_initializer=init_mode, activation='sigmoid')) #binary output layer
    optimizer = RMSprop(learning_rate=0.005)
    model.compile(loss = "binary_crossentropy", optimizer = optimizer, metrics=['accuracy'])
    return model

# create model
model = KerasClassifier(build_fn=create_model, epochs=20, batch_size=100, verbose=0) #model same eah time, pass in kerasclassifier then do parameter grid search

# define the grid search parameters
init_mode = ['uniform', 'lecun_uniform', 'normal', 'zero', 'glorot_normal', 'glorot_uniform', 'he_normal', 'he_uniform']
param_grid = dict(init_mode=init_mode)

# Create Grid Search
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1)
grid_result = grid.fit(x_train, y_train)

# Report Results
print(f"Best: {grid_result.best_score_} using {grid_result.best_params_}")
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print(f"Means: {mean}, Stdev: {stdev} with: {param}") 
```
